```python
import pandas as pandas
import numpy as np
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Create a simple classification dataset
X_sklearn, y_sklearn = make_classification(n_samples=100, n_features=4, n_informative=4, 
                                           n_redundant=0, random_state=42)

# Divide the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_sklearn, y_sklearn, test_size=0.2, random_state=42)

print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

```python
from classificao_padroes.models import KNN

knn = KNN(1, X_train, y_train, minkowskiOrder=2)
y_pred = knn.predict(X_test)
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from classificao_padroes.models import KNN, MDC

mdc = MDC(X_train, y_train, robustVersion=False)
mdc.train()
y_pred = mdc.predict(X_test)

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from classificao_padroes.models import KNN, MDC

mdc = MDC(X_train, y_train, robustVersion=True)
mdc.train()
y_pred = mdc.predict(X_test)

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
